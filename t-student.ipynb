```python
# https://builtin.com/data-science/t-test-python
import numpy as np
from scipy import stats

# Datos de ejemplo
grupo1 = np.array([0.28,0.34,0.34,0.34,0.35,0.35,0.35,0.36,0.36,0.37,0.38,0.40])
grupo2 = np.array([0.49,0.52,0.52,0.52,0.53,0.53,0.53,0.53,0.53,0.54,0.54,0.55])

# Print the variance of both data groups
print(np.var(grupo1), np.var(grupo2), np.var(grupo1) / np.var(grupo2))

# Prueba T de muestras independientes
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretación
alpha = 0.05
if p_value < alpha:
    print("Rechazar la hipótesis nula: Las medias son significativamente diferentes.")
else:
    print("No se rechaza la hipótesis nula: No hay diferencia significativa.")
```
